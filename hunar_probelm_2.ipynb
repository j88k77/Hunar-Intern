```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

# Dataset
data = {
    'Year': [1810, 1820, 1830, 1840, 1850, 1860, 1870, 1880, 1900, 1910, 1920,
             1930, 1940, 1950, 1960, 1970, 1980],
    'Height': [169.7, 169.1, 166.7, 166.5, 165.6, 166.6, 167.2, 168.0, 169.4,
               170.9, 171.0, 173.9, 174.9, 176.0, 176.9, 177.1, 176.8]
}


df = pd.DataFrame(data)


mean_height = df['Height'].mean()
print(f"Mean Height: {mean_height:.2f} cm")


plt.figure(figsize=(10, 6))
plt.plot(df['Year'], df['Height'], marker='o', linestyle='-', color='b')
plt.title('Average Height of Males in the UK (1800s to 1980)', fontsize=14)
plt.xlabel('Year', fontsize=12)
plt.ylabel('Height (cm)', fontsize=12)
plt.grid(True)
plt.axhline(y=mean_height, color='r', linestyle='--', label=f'Mean Height: {mean_height:.2f} cm')
plt.legend()
plt.show()
```
