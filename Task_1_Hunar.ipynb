```python
import numpy as np
import pandas as pd
```

```python
food_dataset = pd.read_csv('/content/food_coded.csv')
```

```python
food_dataset.head()
```

```python
food_dataset.describe()
```

```python
#checking the number of missing values in  the dataset
food_dataset.isnull().sum()
```

```python
food_dataset = food_dataset.loc[:, ~food_dataset.T.duplicated()]

print("Dataset after removing duplicate columns:")
print(food_dataset.head())
```

```python
# Fill missing values for numeric columns with mean
numeric_cols = food_dataset.select_dtypes(include=[np.number]).columns
food_dataset[numeric_cols] = food_dataset[numeric_cols].fillna(food_dataset[numeric_cols].mean())
print("Dataset after filling missing values:")
print(food_dataset.head())
```
